```python
df["Revenue"] = df["Quantity"] * df["Price"]
#Revenue column
```

```python
df[["Quantity", "Price", "Revenue"]].head()
```

```python
df_customers = df[df["Customer ID"].notnull()]
```

```python
df["Revenue"].describe()
```

```python
total_revenue = df["Revenue"].sum()
total_revenue
```

```python
df["InvoiceMonth"] = df["InvoiceDate"].dt.to_period("M")

monthly_revenue = (
    df.groupby("InvoiceMonth")["Revenue"]
      .sum()
      .reset_index()
)

monthly_revenue.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_revenue["InvoiceMonth"].astype(str),
         monthly_revenue["Revenue"])
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
df_positive = df[df["Revenue"] > 0]

```

```python
df_positive["Revenue"].sum()
```

```python
df_positive["InvoiceMonth"] = df_positive["InvoiceDate"].dt.to_period("M")

monthly_revenue_clean = (
    df_positive.groupby("InvoiceMonth")["Revenue"]
               .sum()
               .reset_index()
)

monthly_revenue_clean.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_revenue_clean["InvoiceMonth"].astype(str),
         monthly_revenue_clean["Revenue"])
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend (Excluding Returns)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
purchase_frequency = (
    df_customers.groupby("Customer ID")["Invoice"]
                .nunique()
                .sort_values(ascending=False)
)

purchase_frequency.head(10)
```

## Exploratory Data Analysis Summary

- The dataset contains over 525,000 transaction records, indicating a large and realistic retail dataset.
- Revenue distribution is highly right-skewed, with many small purchases and a few very high-value transactions.
- Negative revenue values were identified, representing product returns and cancellations.
- Approximately a significant portion of revenue is lost due to returns, which were analyzed separately to avoid distortion.
- Monthly revenue trends show clear variation over time, suggesting possible seasonality in customer purchases.
- The United Kingdom contributes the highest share of total revenue.
- A small group of customers accounts for a large proportion of total revenue, indicating high-value repeat customers.
- Customer purchase frequency varies widely, with some customers placing significantly more orders than others.
